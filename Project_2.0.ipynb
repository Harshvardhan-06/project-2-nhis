```python
import os
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
df1 = pd.read_csv("dataset_1 - dataset_1.csv")
df2 = pd.read_csv("dataset_2.xlsx - dataset_2.csv")
df3 = pd.read_csv("dataset_3 - dataset_3.csv")

# Convert 'dteday' to datetime format
df1['dteday'] = pd.to_datetime(df1['dteday'], format='%d-%m-%Y')

# Drop unnecessary columns
if 'Unnamed: 0' in df2.columns:
    df2.drop(columns=['Unnamed: 0'], inplace=True)

# Handle missing values in 'atemp'
df2.loc[:, 'atemp'] = df2['atemp'].fillna(df2['atemp'].median())

# Data Visualization
plt.figure(figsize=(8, 5))
sns.histplot(df1['temp'], bins=30, kde=True, color='blue')
plt.title('Temperature Distribution')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(x=df1['temp'], y=df2['cnt'], alpha=0.5, color='red')
plt.title('Temperature vs Total Rentals')
plt.xlabel('Temperature')
plt.ylabel('Total Rentals')
plt.show()

# Merge datasets into a new dataset named df1.1
df1_1 = pd.merge(df1, df2, on='instant', how='inner')

# Display merged dataset info and first few rows
print(df1_1.info())
print(df1_1.head())

from IPython.display import display  # Import the display function
display (df1_1)  # Display the DataFrame in a table format

# Load datasets
df1_1 = pd.read_csv("df1_1_cleaned.csv")
df3 = pd.read_csv("dataset_3 - dataset_3.csv")

# Convert 'dteday' to datetime format
df1_1['dteday'] = pd.to_datetime(df1_1['dteday'], format='%Y-%m-%d')
df3['dteday'] = pd.to_datetime(df3['dteday'], format='%d-%m-%Y')

# Check and handle missing values in df3
df3.fillna(df3.median(), inplace=True)

# Data Visualization for df1_1
plt.figure(figsize=(8, 5))
sns.histplot(df1_1['cnt'], bins=30, kde=True, color='blue')
plt.title('Total Rentals Distribution in df1_1')
plt.xlabel('Total Rentals')
plt.ylabel('Frequency')
plt.show()

# Data Visualization for df3
plt.figure(figsize=(8, 5))
sns.histplot(df3['cnt'], bins=30, kde=True, color='green')
plt.title('Total Rentals Distribution in df3')
plt.xlabel('Total Rentals')
plt.ylabel('Frequency')
plt.show()

# Merge datasets into a new dataset named df1.2
df1_2 = pd.concat([df1_1, df3], ignore_index=True)

# Save the dataset as a properly formatted CSV file
df1_2.to_csv("df1_2_cleaned.csv", index=False)

# Confirm the file is saved
print("Dataset saved successfully as 'df1_2_cleaned.csv'.")

# Display merged dataset info and first few rows
print(df1_2.info())
print(df1_2.head())

from IPython.display import display  # Import the display function
display (df1_2)  # Display the DataFrame in a table format

# Load dataset
df1_2 = pd.read_csv("df1_2_cleaned.csv")

# Convert 'dteday' to datetime format
df1_2['dteday'] = pd.to_datetime(df1_2['dteday'])

# Check and handle missing values
df1_2.fillna(df1_2.median(), inplace=True)

# Data Visualization
plt.figure(figsize=(8, 5))
sns.histplot(df1_2['cnt'], bins=30, kde=True, color='purple')
plt.title('Total Rentals Distribution in df1_2')
plt.xlabel('Total Rentals')
plt.ylabel('Frequency')
plt.show()

# Boxplot to visualize outliers in numerical columns
numeric_cols = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']
plt.figure(figsize=(12, 6))
df1_2[numeric_cols].boxplot()
plt.title("Boxplot of Numerical Columns (Outlier Detection)")
plt.xticks(rotation=45)
plt.show()

# Handling outliers using IQR method
Q1 = df1_2[numeric_cols].quantile(0.25)
Q3 = df1_2[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

df1_2 = df1_2[~((df1_2[numeric_cols] < (Q1 - 1.5 * IQR)) | (df1_2[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

# Save the cleaned dataset
df1_2.to_csv("df1_2_final_cleaned.csv", index=False)

# Confirm the file is saved
print("Dataset saved successfully as 'df1_2_final_cleaned.csv'.")

# Display final dataset info and first few rows
print(df1_2.info())
print(df1_2.head())
```
